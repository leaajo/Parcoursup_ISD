## **Premier RandomForest : on retire toutes les valeurs manquantes des écarts-types**

```python
lycee_df6 = lycee_df5
lycee_df6 = lycee_df6.dropna(axis=0, subset=["Ecart-type de l'IPS voie PRO"])
lycee_df6.pop("Dont effectif des admis en internat") #à mettre plus haut, pas assez d'infos
lycee_df6.info()
lycee_df6["Dont effectif des admis issus du même établissement (BTS/CPGE)"] = lycee_df6["Dont effectif des admis issus du même établissement (BTS/CPGE)"].replace('nan', np.NaN)
lycee_df6 = lycee_df6.dropna(axis=0, subset=lycee_df6.columns)
lycee_df6.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.inspection import permutation_importance
```

```python

y_admis_lycee = lycee_df6["Dont effectif des admis issus du même établissement (BTS/CPGE)"]
y_admis_lycee_f = lycee_df6["Dont effectif des admises issues du même établissement (BTS/CPGE)"]
y_admis_acad = lycee_df6["Dont effectif des admis issus de la même académie"]
X = lycee_df6.copy()

for col in ["Dont effectif des admis issus du même établissement (BTS/CPGE)", "Dont effectif des admises issues du même établissement (BTS/CPGE)",
            "Dont effectif des admis issus de la même académie", "Dont effectif des admis issus de la même académie (Paris/Créteil/Versailles réunies)"] :
            X.pop(col)
scaler = StandardScaler()
X_n = scaler.fit_transform(X)
```

### **Admis dans le même lycée**

```python
test_size=0.15
X_no_test, X_test, y_no_test, y_test = train_test_split(X_n, y_admis_lycee, test_size=test_size, random_state=3)
X_train, X_val, y_train, y_val = train_test_split(X_no_test, y_no_test, test_size=0.2, random_state=3)
```

```python
X_train.shape
```

```python
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 20],
}

rf_model = RandomForestRegressor()

grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring="r2").fit(X_no_test, y_no_test)

best_params_forest = grid_search.best_params_
best_score_forest = grid_search.best_score_
print("Meilleur paramètre : " + str(best_params_forest))
print("Meilleur score : " + str(best_score_forest))
```

```python
rf_model = RandomForestRegressor(**best_params_forest)
rf_model.fit(X_no_test, y_no_test)
best_score_forest = rf_model.score(X_test, y_test)

print("Meilleur paramètre : " + str(best_params_forest))
print("Meilleur score : " + str(best_score_forest))
```

```python
result = permutation_importance(
    rf_model, X_no_test, y_no_test, n_repeats=10, random_state=2, scoring='r2')

importance_var = pd.Series(result.importances_mean, index=X.columns.to_list())
```

```python
fig, ax = plt.subplots()
importance_var.plot.bar(yerr=result.importances_std, ax=ax)
ax.set_title("Importance des variables déterminée par permutation")
ax.set_ylabel("Diminution moyenne du score R2")
fig.tight_layout()
plt.show()
```
